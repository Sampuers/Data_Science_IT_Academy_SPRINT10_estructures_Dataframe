#### Voy a utilizar el primer método de raspado  *_BeautifulSoup_* 


```python
import requests

URL = "http://quotes.toscrape.com"
page = requests.get(URL)

print(page.text)
```

Los datos parecen un gran lío, presentan atributos dispersos. Agregando las nuevas líneas de código, se crea un objeto que se toma como entrada. 

```python
from bs4 import BeautifulSoup

URL = "http://quotes.toscrape.com"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

Ya se muestra de forma más ordenada y ahora vamos a filtrar por mis intereses, que serán el autor y las frases de la primera página. 

```python
import pandas as pd
```

```python
# Encontrar todas las etiquetas <span> con la clase "text"
etiquetas = soup.find_all("span", class_="text")

# Listas para almacenar datos
textos = []
autores = []

# Iterar sobre las etiquetas de las citas
for etiqueta in etiquetas:
    # Extraer texto de la cita
    texto_etiqueta = etiqueta.text.strip()

    # Encontrar la etiqueta <small> con la clase "author" dentro del contenedor de la cita
    autor = etiqueta.find_next("small", class_="author")
    # Extraer texto del autor
    autor_text = autor.text.strip()

    # Almacenar en listas
    textos.append(texto_etiqueta)
    autores.append(autor_text)

    # Imprimir los resultados como desees
    print(f"Texto: {texto_etiqueta}")
    print(f"Autor: {autor_text}")
    print("------------------")

# Crear DataFrame
df = pd.DataFrame({'Texto': textos, 'Autor': autores})

# Imprimir el DataFrame
df
```

- Voy a configurar el archivo de setting para ignorar la reestricción en el uso de Robots. En el archivo que se genera en Home, entro en seetings para cambiar de true a false. 

- Se hace Scraping de la página web. 

```python
import scrapy
from scrapy.crawler import CrawlerProcess

class CraigDavidSpider(scrapy.Spider):
    name = 'craigdavid'
    start_urls = ['https://www.last.fm/es/search/tracks?q=Craig+David']

    def parse(self, response):
        result_list = []

        for track in response.xpath('//*[@id="mantle_skin"]/div[3]/div/div[1]/table/tbody/tr'):
            # Extraer título y artista usando XPath
            title = track.xpath('td[4]/a/text()').get()
            artist = 'Craig David'  # Establecer el artista manualmente

            # Agregar a la lista
            if title:
                result_list.append({'title': title, 'artist': artist})

        return result_list

# Configurar el proceso de rastreo con salida a un archivo JSON
process = CrawlerProcess(settings={
    'FEEDS': {
        'craigdavid_tracks.json': {
            'format': 'json',
            'encoding': 'utf8',
        },
    },
    'USER_AGENT': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
})

# Iniciar el proceso de rastreo
process.crawl(CraigDavidSpider)
process.start()
```
